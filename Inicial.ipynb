### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado


Referência do pyautogui: https://pyautogui.readthedocs.io/en/latest/quickstart.html

```python
# Passo á passo
# Será instalado o pyautogui (automação) e pandas (dados)
# Em resumo, o pyautogui faz os periféricos se moverem e o pandas insere as informações
# E adicionando outro complemente, será instalado o openpyxl (que faz o pandas ler a base de dados em excel, ou seja, arquivos em .xlsx)
# Para inseri-los, basta clicar em Terminal > New Terminal. Irá abrir um terminal abaixo e note que há um (+) e ao lado uma seta para baixo, clique na mesma e clique em Command Prompt
# No terminal, digite: pip install pyautogui pandas openpyxl. Após isso, clique em Enter e aguarde a instalação

# Passo 1 - Entrar no sistema da empresa (link do drive)
# Abra o navegador que a empresa utiliza, no meu caso, o Google Chrome
# Digite o link da empresa
# Aperte enter e espere
# Passo 2 - Ao navegar no sistema, encontrar a base de dados
# Passo 3 - Exportar a base de dados (exportar o arquivo)
# Passo 4 - Calcular os indicadores (faturamento e quantidade de produtos vendidos)
# Passo 5 - Enviar as informações para o e-mail

# Será instalado o pyautogui (automação) e pandas (dados)
# Em resumo, o pyautogui faz os periféricos se moverem e o pandas insere as informações
# E adicionando outro complemente, será instalado o openpyxl (que faz o pandas ler a base de dados em excel, ou seja, arquivos em .xlsx)
# Para inseri-los, basta clicar em Terminal > New Terminal. Irá abrir um terminal abaixo e note que há um (+) e ao lado uma seta para baixo, clique na mesma e clique em Command Prompt
# No terminal, digite: pip install pyautogui pandas openpyxl. Após isso, clique em Enter e aguarde a instalação

```

```python
import pyautogui # Importação da biblioteca
import time # Importação de biblioteca para controlar o tempo
pyautogui.PAUSE = 0.5 # Configura o tempo de pausa entre os comandos do pyautogui, ou seja, a cada comando, ele irá esperar 0.5 segundos para executar o próximo comando. Isso é importante para evitar que o programa execute os comandos muito rápido e cause erros
# Passo 1 - Entrar no sistema da empresa (link do drive)
# Abra o navegador que a empresa utiliza, no meu caso, o Google Chrome
pyautogui.press("win") # Pressiona a tecla Windows do teclado
pyautogui.write("chrome") # Escreve chrome no teclado
pyautogui.press("enter") # Irá apertar o enter do teclado
time.sleep(2) # Irá esperar 2 segundos para inserir o próximo passo
# Digite o link da empresa
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga" # Criei essa variável para facilitar a escrita do link, caso seja necessário alterar o link no futuro, basta alterar essa variável
pyautogui.write(link) # Irá escrever o link da empresa inserido
pyautogui.press("enter") # Irá apertar o enter do teclado
time.sleep(3) # Irá esperar 3 segundos para inserir o próximo passo 
```

```python
# Passo 2 - Ao navegar no sistema, encontrar a base de dados
pyautogui.click(x=442, y=519, clicks=2) # Isso é o resultado do comando acima, ou seja, as coordenadas do mouse. O comando click é para clicar, o x é a coordenada x do mouse, o y é a coordenada y do mouse e o clicks é para clicar duas vezes
time.sleep(2) # Irá esperar 2 segundos para inserir o próximo passo
# Passo 3 - Exportar a base de dados (exportar o arquivo)
pyautogui.click(clicks=1) # Seleciona o arquivo
time.sleep(2) # Irá esperar 2 segundos para inserir o próximo passo
pyautogui.click(x=602, y=522,clicks=1) # Ele segue essa coordenada X e Y e clica nos 3 pontinhos
time.sleep(2) # Irá esperar 2 segundos para inserir o próximo passo
pyautogui.click(x=663, y=625, clicks=1) # Clica para fazer o download do arquivo
time.sleep(2) # Irá esperar 2 segundos para inserir o próximo passo
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas # Importação da biblioteca para manipulação dos dados
# Passo 4 - Calcular os indicadores (faturamento e quantidade de produtos vendidos)
# Abrir a base de dados
caminho = r"D:\Arquivos Usuário\Downloads\Vendas - Dez.xlsx" # Criei essa variável para indicar o caminho do arquivo baixado, o r antes da string é para indicar que é uma raw string, ou seja, uma string literal, isso é importante para evitar erros com as barras invertidas do caminho do arquivo
tabela = pandas.read_excel(caminho) # O comando read_excel é para ler o arquivo excel, o caminho é o caminho do arquivo que criamos na variável acima
# Ver as informações da base de dados
display(tabela) # O comando display é para mostrar a tabela no terminal. Ele é diferente do print, pois o print mostra a tabela de forma desformatada, já o display mostra a tabela de forma formatada, ou seja, com as colunas e linhas organizadas.
# Somar o faturamento de todos os produtos = somar a coluna de valor final
faturamento = tabela["Valor Final"].sum() # O comando sum é para somar os valores da coluna "Valor Final", ou seja, o faturamento total
# Somar a quantidade de produtos
quantidade_produtos = tabela["Quantidade"].sum() # O comando sum é para somar os valores da coluna "Quantidade", ou seja, a quantidade total de produtos vendidos
print(faturamento) # Imprime o faturamento total
print(quantidade_produtos) # Imprime a quantidade total de produtos vendidos
```

### Vamos agora enviar um e-mail pelo gmail

```python
import pyperclip # Importação da biblioteca para manipulação da área de transferência, ou seja, para copiar e colar textos
# Passo 5 - Enviar as informações para o e-mail
# Abrir uma nova aba
pyautogui.hotkey("ctrl", "t") # O comando hotkey é para apertar as teclas ao mesmo tempo, nesse caso, o ctrl e o t, que é para abrir uma nova aba no navegador
# Entrar no e-mail
pyautogui.write("https://mail.google.com/") # Escreve o link do gmail no teclado
pyautogui.press("enter") # Irá apertar o enter do teclado
time.sleep(10) # Irá esperar 10 segundos para inserir o próximo passo
# Clicar no botão Escrever
pyautogui.click(x=221, y=238, clicks=1) # Ele segue essa coordenada X e Y e clica para escrever um novo e-mail
time.sleep(3) # Irá esperar 3 segundos para inserir o próximo passo
# Para quem enviar o e-mail
pyautogui.write("mtsfaracini21@gmail.com" , interval=0.1) # Escreve o e-mail do destinatário no teclado, o interval é para controlar a velocidade da escrita, ou seja, a cada letra, ele irá esperar 0.1 segundos para escrever a próxima letra. Isso é importante para evitar que o programa escreva muito rápido e cause erros
time.sleep(1) # Irá esperar 1 segundo para inserir o próximo passo
pyautogui.press("enter") # Irá apertar o enter do teclado
time.sleep(1) # Irá esperar 1 segundo para inserir o próximo passo
pyautogui.press("tab") # Seleciona o e-mail digitado acima
time.sleep(1)
# O assunto do e-mail
pyperclip.copy("Relatório de Vendas") # O comando copy é para copiar o texto "Relatório de Vendas" para a área de transferência, ou seja, para o ctrl + c do teclado. Isso é importante para evitar erros de digitação e para facilitar a escrita do assunto do e-mail  
pyautogui.hotkey("ctrl", "v") # O comando hotkey é para apertar as teclas ao mesmo tempo, nesse caso, o ctrl e o v, que é para colar o texto copiado acima no assunto do e-mail   
pyautogui.press("tab") # Seleciona para o campo de corpo do e-mail
# O corpo do e-mail
texto = f"""
Prezados(as)
Segue o relatório de vendas de hoje.
Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {quantidade_produtos:,}
Qualquer dúvida, estou á disposição.
Atenciosamente:
"""
pyperclip.copy(texto) # O comando copy é para copiar o texto do corpo do e-mail para a área de transferência, ou seja, para o ctrl + c do teclado. Isso é importante para evitar erros de digitação e para facilitar a escrita do corpo do e-mail
pyautogui.hotkey("ctrl", "v") # O comando hotkey é para apertar as teclas ao mesmo tempo, nesse caso, o ctrl e o v, que é para colar o texto copiado acima no corpo do e-mail
# Enviar o e-mail
pyautogui.click(x=1201, y=971, clicks=1) # Ele segue essa coordenada X e Y e clica para enviar o e-mail
```
